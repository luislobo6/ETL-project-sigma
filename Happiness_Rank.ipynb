```python
import pandas as pd
```

```python
file2015 = "Resources/Happiness/2015.csv"
df15 = pd.read_csv(file2015)
df15_n=df15[["Country","Happiness Rank"]]
df15_n.head()
```

```python
file2016 = "Resources/Happiness/2016.csv"
df16 = pd.read_csv(file2016)
df16_n=df16[["Country","Happiness Rank"]]
df16_n.head()
```

```python
file2017 = "Resources/Happiness/2017.csv"
df17 = pd.read_csv(file2017)
df17_n=df17[["Country","Happiness.Rank"]]
df17_n=df17_n.rename(columns={"Happiness.Rank":"Happiness Rank"})
df17_n.head()
```

```python
file2018 = "Resources/Happiness/2018.csv"
df18 = pd.read_csv(file2018)
df18_n=df18[["Country or region","Score"]]
df18_n=df18_n.rename(columns={"Country or region":"Country",
                             "Score":"Happiness Rank"})
df18_n.head()
```

```python
file2019 = "Resources/Happiness/2019.csv"
df19 = pd.read_csv(file2019)
df19_n=df19[["Country or region","Score"]]
df19_n=df19_n.rename(columns={"Country or region":"Country",
                             "Score":"Happiness Rank"})
df19_n.head()
```

```python
dfA=df15_n.merge(df16_n, how="inner", on="Country", suffixes=("_15","_16"))
dfB=dfA.merge(df17_n, how="inner", on="Country")
dfC=dfB.merge(df18_n, how="inner", on="Country",suffixes=("_17","_18"))
df_Tot=dfC.merge(df19_n,how="inner", on="Country")
df_Tot=df_Tot.rename(columns={"Happiness Rank":"Happiness Rank_19"})
df_Tot.head()
```
